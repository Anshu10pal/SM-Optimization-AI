COMBINE EVERYTHING

```python
final_df = pd.DataFrame({
    "res_conf": res_conf,
    "sub_conf": sub_conf,
    "fault_conf": fault_conf,
    "res_correct": res_correct,
    "sub_correct": sub_correct,
    "fault_correct": fault_correct
})

final_df["all_correct"] = (
    final_df["res_correct"] &
    final_df["sub_correct"] &
    final_df["fault_correct"]
)
```

FINAL AUTO DECISION LOGIC

```python
auto_df = final_df[
    (final_df["res_conf"] >= 0.75) &
    (final_df["sub_conf"] >= 0.60) &
    (final_df["fault_conf"] >= 0.80)
]

print("AUTO accuracy:",
      round(auto_df["all_correct"].mean()*100,2), "%")

print("AUTO coverage:",
      round(len(auto_df)/len(final_df)*100,2), "%")
```
